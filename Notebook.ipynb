Ya cargado el dataset, se procede a realizar un análisis exploratorio y junto a ello la limpieza de datos (valores) para garantizar que los datos estén limpios y listos para el análisis.
```python
# Verifica valores nulos (NaN) en cada columna
valores_nulos_por_columna = df.isnull().sum()

# Verifica valores vacíos en cada columna
valores_vacios_por_columna = (df == '').sum()

# Imprime el resultado
print("Valores nulos (NaN) por columna:")
print(valores_nulos_por_columna)

print("\nValores vacíos por columna:")
print(valores_vacios_por_columna)
```

Este código verifica la cantidad de valores nulos (NaN) y valores vacíos en cada columna del DataFrame "df" y luego imprime los resultados por separado. La variable "valores_nulos_por_columna" almacena la suma de valores nulos en cada columna, mientras que "valores_vacios_por_columna" almacena la cantidad de valores vacíos (representados como cadenas vacías) en cada columna del DataFrame. Las cantidades se imprimen en dos secciones separadas para una inspección rápida de la calidad de los datos.

```python
# Crea la carpeta 'data' si no existe
if not os.path.exists('data'):
    os.makedirs('data')

# Columnas a exportar
columnas_a_exportar = df.columns

# Estilos para el párrafo en el PDF
styles = getSampleStyleSheet()

for columna in columnas_a_exportar:
    # Obtiene la lista completa de valores únicos en la columna actual
    valores_unicos = df[columna].unique()

    # Especifica el nombre del archivo PDF de salida basado en el nombre de la columna
    nombre_archivo = f'data/valores_unicos_{columna}.pdf'

    # Configura el tamaño del papel para el PDF (por ejemplo, letter)
    pdf = SimpleDocTemplate(nombre_archivo, pagesize=letter)

    # Lista para almacenar los elementos del PDF
    elements = []

    # Agrega el título de la columna al PDF sin aplicar ningún estilo adicional
    elements.append(Paragraph(f"Valores únicos en la columna '{columna}':", styles["Title"]))
    elements.append(Spacer(1, 12))  # Agrega un espacio en blanco después del título

    # Agrega cada valor único en una línea separada sin aplicar ningún estilo adicional
    for valor in valores_unicos:
        elements.append(Paragraph(str(valor), styles["Normal"]))

    # Crea el PDF con los elementos
    pdf.build(elements)

    print(f"Los valores únicos en la columna '{columna}' se han guardado en el archivo {nombre_archivo}")
```

Este código primero verifica si la carpeta 'data' existe; de lo contrario, la crea. Luego, itera a través de las columnas del DataFrame "df" y genera un archivo PDF separado para cada columna. Estos archivos contienen una lista de valores únicos en esa columna. Se utiliza ReportLab para crear los PDF, y los nombres de archivo se basan en el nombre de la columna. Finalmente, se imprime un mensaje que informa que los valores únicos en cada columna se han guardado como archivos PDF en la carpeta 'data'.

```python
# Aplica unidecode y convierte a minúsculas todas las celdas del DataFrame
df = df.applymap(lambda x: unidecode(str(x)).lower())

# Convierte los títulos de las columnas a minúsculas (opcional)
df.columns = df.columns.str.lower()

# Aplica una expresión regular para eliminar espacios y contenido entre paréntesis en la columna 'municipio'
df['municipio'] = df['municipio'].str.replace(r'\s+\([^)]*\)', '', regex=True)

# Aplica str.replace para quitar "d.c." y el espacio que lo precede de la columna 'municipio'
df['municipio'] = df['municipio'].str.replace(r'\s+d\.c\.$', '', regex=True)

# Reemplaza los espacios en los nombres de las columnas por guiones bajos (_)
df.columns = df.columns.str.replace(' ', '_')

# Reemplaza los valores que contienen '-' (incluso si están rodeados de espacios) en la columna 'armas_medios' por 'no reportado'
df['armas_medios'] = df['armas_medios'].replace(r'\s*-\s*', 'no reportado', regex=True)

# Reemplaza los valores que contienen '-' (incluso si están rodeados de espacios) en la columna 'genero' por 'no reporta'
df['genero'] = df['genero'].replace(r'\s*-\s*', 'no reporta', regex=True)

# Reemplaza los valores que contienen '-' (incluso si están rodeados de espacios) en la columna 'grupo_etario' por 'no reporta'
df['grupo_etario'] = df['grupo_etario'].replace(r'\s*-\s*', 'no reporta', regex=True)

# Para verificar el resultado, puedes imprimir el DataFrame
df
```

```python
# Filtra los datos para eliminar los sectores con un porcentaje del 0%
count_tipo_armas = df['tipo_de_arma'].value_counts()
porcentajes = (count_tipo_armas / count_tipo_armas.sum() * 100).round(1)
count_tipo_armas = count_tipo_armas[porcentajes > 0]
porcentajes = porcentajes[porcentajes > 0]

# Crea un gráfico de pastel con los datos filtrados
plt.figure(figsize=(8, 8))
wedges, _, autopcts = plt.pie(count_tipo_armas, labels=[''] * len(count_tipo_armas), autopct='', startangle=140)
plt.axis('equal')
plt.title('Distribución de Tipos de Armas')

# Personaliza las etiquetas de la leyenda con nombres y porcentajes
legend_labels = [f'{i + 1}. {label} ({porcentaje}%)' for i, (label, porcentaje) in enumerate(zip(count_tipo_armas.index, porcentajes))]

# Ubica la leyenda en la esquina inferior derecha pero separada del gráfico
legend = plt.legend(legend_labels, title='Tipos de Armas', bbox_to_anchor=(1.05, 0.5), loc='center left')

# Agrega números en cada sector del gráfico
for i, (wedge, label) in enumerate(zip(wedges, legend_labels)):
    angle = wedge.theta1 + 0.5 * (wedge.theta2 - wedge.theta1)
    x = 0.7 * wedge.r * np.cos(np.deg2rad(angle))
    y = 0.7 * wedge.r * np.sin(np.deg2rad(angle))
    plt.text(x, y, str(i + 1), ha='center', va='center', fontsize=12, fontweight='bold', color='white')

# Agrega números dentro del cuadro de la leyenda
for i, (text, label) in enumerate(zip(legend.get_texts(), legend_labels)):
    text.set_text(label)

plt.show()
```

```python
# Define Panel widgets
year_slider = pn.widgets.IntRangeSlider(name='Rango de años', width=250, start=2010, end=2023, value=(2010, 2023), value_throttled=(2010,2023))
year_slider
```

```python
# Crear botones de selección de tipo de datos
bt2 = pn.widgets.RadioButtonGroup(
    name='Seleccionar datos',
    button_type='success',
    options=['delito', 'tipo_de_arma'],
)

bt2
```

```python
# Radio buttons para seleccionar entre 'departamento' y 'municipio'
radio_buttons = pn.widgets.RadioButtonGroup(
    name='Seleccionar datos',
    options=['departamento', 'municipio'],
    button_type='success'
)

radio_buttons 
```

```python
# Función para crear la gráfica interactiva
@pn.depends(year=year_slider.param.value_throttled, data_type=radio_buttons) 
def create_bar_chart(year, data_type):
    # Filtra el DataFrame por el rango de años seleccionado
    filtered_df = df[(df['year'] >= year[0]) & (df['year'] <= year[1])]
    
    # Calcula la frecuencia de departamentos o municipios según la selección
    if data_type == 'departamento':
        data_column = 'departamento'
        plot_title = 'Frecuencia de Departamentos (Top 32)'
    else:
        data_column = 'municipio'
        plot_title = 'Frecuencia de Municipios (Top 32)'
    
    frecuencia_data = filtered_df[data_column].value_counts().reset_index()
    frecuencia_data = frecuencia_data.rename(columns={'index': data_column, data_column: 'Frecuencia'})
    
    # Selecciona los 32 principales
    top_32 = frecuencia_data.nlargest(32, 'Frecuencia')
    
    # Crea el gráfico de barras
    bar_chart = top_32.hvplot(
        kind='bar',
        x=data_column, 
        y='Frecuencia', 
        xlabel=data_column, 
        ylabel='Frecuencia', rot=90,
        title=plot_title
    )
    
    return bar_chart

# Crea un panel con los widgets y el gráfico
plot = pn.Column(year_slider, radio_buttons, create_bar_chart)

plot
```

```python
# Función para crear la tabla paginada
@pn.depends(data_type=radio_buttons.param.value, year=year_slider.param.value)
def create_table(data_type, year):
    # Filtrar el DataFrame por el rango de años seleccionado
    filtered_df = df[(df['year'] >= year[0]) & (df['year'] <= year[1])]

    # Agrupar los datos por departamento o municipio según la selección
    if data_type == 'departamento':
        data_column = 'departamento'
    else:
        data_column = 'municipio'

    grouped_data = filtered_df.groupby(['year', data_column]).size().reset_index(name='Frecuencia')

    # Ordenar los datos por frecuencia de mayor a menor
    grouped_data = grouped_data.sort_values(by='Frecuencia', ascending=False)

    # Crear una tabla interactiva paginada, no editable
    table = pn.widgets.Tabulator(grouped_data, pagination='remote', page_size=10)
    
    # Configurar las columnas y hacer que la tabla no sea editable
    table.param.columns = [data_column, 'Frecuencia']
    table.param.editable = False

    return table

# Crear un panel con el deslizador de años, los botones y la tabla interactiva
panel = pn.Column(
    year_slider,
    radio_buttons,
    create_table
)

panel
```

```python
# Función para crear la tabla paginada
@pn.depends(data_type=bt2.param.value, year=year_slider.param.value)
def create_table2(data_type, year):
    # Filtrar el DataFrame por el rango de años seleccionado
    filtered_df = df[(df['year'] >= year[0]) & (df['year'] <= year[1])]

    # Agrupar los datos por delitos o tipo_de_arma según la selección
    if data_type == 'delito':
        data_column = 'delito'
    else:
        data_column = 'tipo_de_arma'

    grouped_data = filtered_df.groupby(['year', data_column]).size().reset_index(name='Frecuencia')

    # Ordenar los datos por frecuencia de mayor a menor
    grouped_data = grouped_data.sort_values(by='Frecuencia', ascending=False)

    # Crear una tabla interactiva paginada
    table2 = pn.widgets.Tabulator(grouped_data, pagination='remote', page_size=10)
    
    # Configurar las columnas y hacer que la tabla no sea editable
    table2.columns = [data_column, 'Frecuencia']
    table2.editable = False

    return table2

# Crear un panel con el deslizador de años, los botones y la tabla interactiva
panel2 = pn.Column(
    year_slider,
    bt2,
    create_table2
)

panel2
```
